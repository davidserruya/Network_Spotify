## B - Version "without no successors nodes"

```python
less_tranding_network_restricted = nx.DiGraph()

# Iterate over all nodes in less_tranding_network
for node in less_tranding_network.nodes():
    # Check if the node has at least one successor
    if less_tranding_network.out_degree(node) > 0:
        # Add the node to less_tranding_network_restricted
        less_tranding_network_restricted.add_node(node)
        
        # Also add all outgoing edges from this node to less_tranding_network_restricted
        for succ in less_tranding_network.successors(node):
            # Ensure that the successor is also a node with successors before adding
            if less_tranding_network.out_degree(succ) > 0:
                less_tranding_network_restricted.add_edge(node, succ)
```

```python
# plot superstar_network
nx.draw(less_tranding_network_restricted)
plt.show()
```

```python
# Calculate the degrees (sum of in-degree and out-degree) of each node
degrees = {node: val for (node, val) in less_tranding_network_restricted.degree()}

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(less_tranding_network_restricted, with_labels=True, node_size=[v * 10 for v in degrees.values()], 
        node_color='skyblue', font_weight='bold', font_size=5, 
        arrowstyle='-|>', arrowsize=8)
plt.title("less_tranding_network_restricted_Graph")
plt.show()
```

### Centrality

```python
# The graph is directed so we can use in and out degrees

# indegree centrality (incoming ties, only for directed graphs)
print(nx.in_degree_centrality(less_tranding_network_restricted))

# outdegree centrality (outgoing ties)
print(nx.out_degree_centrality(less_tranding_network_restricted))

# degree centrality (sum of in and out degrees)
print(nx.degree_centrality(less_tranding_network_restricted))
```
